```python
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("station_with_bikes.csv",header=0,index_col=0)
```

```python
data.time = pd.to_datetime(data.time)
data.date = pd.to_datetime(data.date)
data
```

```python
data['day_of_year'] = data['date'].dt.dayofyear
data
```

```python
data['time'] = data['time'].dt.strftime('%H:%M')
data
```

```python
time_convert = pd.DataFrame(pd.date_range(start="00:00:00",end="23:55:00",freq ='5T').strftime('%H:%M'))
time_convert.rename(columns = {0:'time'}, inplace = True)
time_convert['to_int'] = pd.RangeIndex(start=1,stop=time_convert.shape[0]+1)
time_convert = time_convert.set_index('time')
time_convert
```

```python
print(data)
data2 = data.join(time_convert,on="time")
data2
```

```python
#problem with this line
data2.loc[data2['to_int'].isnull()==True]
#For now I remove it
data2 = data2.drop(index=8345269)
#No more problem
data2.loc[data2['to_int'].isnull()==True]
```

```python
data2['day_week'] = data['date'].dt.dayofweek
data2
```

```python
from sklearn.neural_network import MLPRegressor

X = data2[['to_int','day_week','day_of_year']].values.reshape(-1, 3)
y = data2['bikesAvailable'].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

model = MLPRegressor(random_state = 1)
model.fit(X_train,y_train)

print(model.predict(X_test))
print(y_test)
model.score(X_test,y_test)
```
