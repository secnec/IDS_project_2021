```python
import pandas as pd
from sklearn.neural_network import MLPRegressor
from datetime import datetime, timedelta
```

```python
#I take a random data point
date = datetime(2019,9,23,11,45)
print(date)
```

```python
#6 hours after
date_6 = date + timedelta(hours=6)
print(date_6)
```

```python
#I load the data and take all the data before my chosen date
bike_arrivals = pd.read_csv("final_arrivals_data.csv",header=0,index_col=0)
```

```python
data = bike_arrivals.loc[:bike_arrivals.loc[pd.to_datetime(bike_arrivals['Timestamp'])==date].index[-1]]
data
```

```python
# The next 6 hours that we want to predict
to_predict_arrivals = bike_arrivals.loc[bike_arrivals.loc[pd.to_datetime(bike_arrivals['Timestamp'])==date].index[-1]+1:bike_arrivals.loc[pd.to_datetime(bike_arrivals['Timestamp'])==date_6].index[-1]]
to_predict_arrivals
```

```python
X = data[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8)
y = data['arrivals'].values.reshape(-1, 1).ravel()

model = MLPRegressor(random_state = 1)
model.fit(X,y)

print(model.predict(to_predict_arrivals[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8)))
print(to_predict_arrivals['arrivals'].values.reshape(-1, 1).ravel())
model.score(to_predict_arrivals[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8),to_predict_arrivals['arrivals'].values.reshape(-1, 1).ravel())
```

```python
predict_values = model.predict(to_predict_arrivals[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8))
round_predict_arrivals = [ round(elem) for elem in predict_values ]
round_predict_arrivals
```

```python
sum(abs(a-b) == 0 for a,b in zip(round_predict_arrivals,list(to_predict_arrivals['arrivals'].values.reshape(-1, 1).ravel().reshape(1, -1)[0])))/len(list(to_predict_arrivals['arrivals'].values.reshape(-1, 1).ravel().reshape(1, -1)[0]))
```

```python
sum(abs(a-b) <= 1 for a,b in zip(round_predict_arrivals,list(to_predict_arrivals['arrivals'].values.reshape(-1, 1).ravel().reshape(1, -1)[0])))/len(list(to_predict_arrivals['arrivals'].values.reshape(-1, 1).ravel().reshape(1, -1)[0]))
```

```python
bike_departures = pd.read_csv("final_departures_data.csv",header=0,index_col=0)
data = bike_departures.loc[:bike_departures.loc[pd.to_datetime(bike_departures['Timestamp'])==date].index[-1]]
X = data[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8)
y = data['departures'].values.reshape(-1, 1).ravel()

model = MLPRegressor(random_state = 1)
model.fit(X,y)
```

```python
to_predict_departures = bike_departures.loc[bike_departures.loc[pd.to_datetime(bike_departures['Timestamp'])==date].index[-1]+1:bike_departures.loc[pd.to_datetime(bike_departures['Timestamp'])==date_6].index[-1]]
print(model.predict(to_predict_departures[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8)))
print(to_predict_departures['departures'].values.reshape(-1, 1).ravel())
```

```python
predict_values = model.predict(to_predict_departures[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8))
round_predict_departures = [ round(elem) for elem in predict_values ]
print(sum(abs(a-b) == 0 for a,b in zip(round_predict_departures,list(to_predict_departures['departures'].values.reshape(-1, 1).ravel().reshape(1, -1)[0])))/len(list(to_predict_departures['departures'].values.reshape(-1, 1).ravel().reshape(1, -1)[0])))
print(sum(abs(a-b) <= 1 for a,b in zip(round_predict_departures,list(to_predict_departures['departures'].values.reshape(-1, 1).ravel().reshape(1, -1)[0])))/len(list(to_predict_departures['departures'].values.reshape(-1, 1).ravel().reshape(1, -1)[0])))
```

```python
to_predict_arrivals[['id','Timestamp']].assign(predict_arrivals=round_predict_arrivals)
```

```python
to_predict_departures[['id','Timestamp']].assign(predict_departures=round_predict_departures)
```

```python
predict_df = to_predict_arrivals[['id','Timestamp']].assign(predict_arrivals=round_predict_arrivals).merge(to_predict_departures[['id','Timestamp']].assign(predict_departures=round_predict_departures), how='outer', on=['id','Timestamp'])
predict_df = predict_df.fillna(0)
predict_df['bikes_evolution'] = predict_df['predict_arrivals'] - predict_df['predict_departures']
predict_df
```

```python
bikes = pd.read_csv('station_with_bikes.csv',header=0,index_col=0)
bikes
```

```python
bikes = bikes.loc[(bikes['date']=='2019/09/23') & (bikes['time'].str.contains('11:45:'))]
bikes['Timestamp'] = pd.to_datetime(bikes['date'].astype(str) + bikes['time'].astype(str), format = '%Y/%m/%d%H:%M:%S').dt.strftime("%Y-%m-%d %H:%M:00")
bikes.drop(columns=['time','date','state'],inplace=True)
bikes
```

```python
bikes['bikes_evolution'] = bikes['bikesAvailable']
bikes['spaces_evolution'] = bikes['spacesAvailable']
bikes.drop(columns=['bikesAvailable','spacesAvailable'],inplace=True)
bikes
```

```python
predict_df['spaces_evolution'] = -1 * predict_df['bikes_evolution']
predict_df.drop(columns=['predict_arrivals','predict_departures'],inplace=True)
predict_df
```

```python
evolution_df = pd.concat([predict_df,bikes])
evolution = evolution_df.sort_values(by='Timestamp')
evolution
```

```python
ev2 = evolution.groupby(['id']).cumsum()
ev2
```

```python
#It worked !!
ev2 = ev2.assign(id=evolution['id'])
ev2 = ev2.assign(Timestamp = evolution['Timestamp'])
ev2.loc[ev2['id']==751]
```
