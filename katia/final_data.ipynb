```python
import pandas as pd
bike_arrivals = pd.read_csv("final_arrivals_data.csv",header=0,index_col=0)
bike_departures = pd.read_csv("final_departures_data.csv",header=0,index_col=0)
```

```python
bike_arrivals
```

```python
bike_departures
```

```python
time_convert = pd.DataFrame(pd.date_range(start="00:00:00",end="23:55:00",freq ='5T').strftime('%H:%M:%S'))
time_convert.rename(columns = {0:'time'}, inplace = True)
time_convert['to_int'] = pd.RangeIndex(start=1,stop=time_convert.shape[0]+1)
time_convert = time_convert.set_index('time')
time_convert
```

```python
bike_arrivals['Timestamp'] = pd.to_datetime(bike_arrivals['Timestamp'])
bike_arrivals['day_of_year'] = bike_arrivals['Timestamp'].dt.dayofyear
bike_arrivals['day_of_week'] = bike_arrivals['Timestamp'].dt.dayofweek
bike_arrivals['time'] = bike_arrivals['Timestamp'].dt.strftime('%H:%M:%S')
bike_arrivals
```

```python
bike_arrivals = bike_arrivals.join(time_convert,on="time")
bike_arrivals.drop(columns="time",inplace=True)
bike_arrivals.rename(columns = {'to_int':'time'}, inplace = True)
bike_arrivals
```

```python
bike_departures['Timestamp'] = pd.to_datetime(bike_departures['Timestamp'])
bike_departures['day_of_year'] = bike_departures['Timestamp'].dt.dayofyear
bike_departures['day_of_week'] = bike_departures['Timestamp'].dt.dayofweek
bike_departures['time'] = bike_departures['Timestamp'].dt.strftime('%H:%M:%S')
bike_departures = bike_departures.join(time_convert,on="time")
bike_departures.drop(columns="time",inplace=True)
bike_departures.rename(columns = {'to_int':'time'}, inplace = True)
bike_departures
```

```python
bike_arrivals.to_csv('final_arrivals_data.csv')
bike_departures.to_csv('final_departures_data.csv')
```
