```python
import pandas as pd
```

```python
bike_arrivals = pd.read_json("arrivals.json")
bike_departures = pd.read_json("departures.json")
```

```python
time_convert = pd.DataFrame(pd.date_range(start="00:00:00",end="23:55:00",freq ='5T').strftime('%H:%M:%S'))
time_convert.rename(columns = {0:'time'}, inplace = True)
time_convert['to_int'] = pd.RangeIndex(start=1,stop=time_convert.shape[0]+1)
time_convert = time_convert.set_index('time')
time_convert
```

```python
bike_arrivals['day_of_year'] = bike_arrivals['datetime'].dt.dayofyear
bike_arrivals = bike_arrivals.join(time_convert,on="time")
bike_arrivals.drop(columns="time",inplace=True)
bike_arrivals.rename(columns = {'to_int':'time'}, inplace = True)
bike_arrivals
```

```python
bike_departures['day_of_year'] = bike_departures['datetime'].dt.dayofyear
bike_departures = bike_departures.join(time_convert,on="time")
bike_departures.drop(columns="time",inplace=True)
bike_departures.rename(columns = {'to_int':'time'}, inplace = True)
bike_departures
```

```python
bike_arrivals.to_json("clean_arivals.json")
bike_departures.to_json("clean_departures.json")
```
