```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```python
bike_arrivals = pd.read_csv("final_arrivals_data.csv",header=0,index_col=0)
bike_arrivals
```

```python
X = bike_arrivals[['time','day_of_week','day_of_year','Cloud amount (1/8)','Relative humidity (%)','Precipitation intensity (mm/h)','Air temperature (degC)','Wind speed (m/s)']].values.reshape(-1, 8)
y = bike_arrivals['arrivals'].values.reshape(-1, 1).ravel()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

model = MLPRegressor(random_state = 1)
model.fit(X_train,y_train)

print(model.predict(X_test))
print(y_test)
model.score(X_test,y_test)
```

```python
sum(abs(a-b) <=1 for a,b in zip(model.predict(X_test),list(y_test.reshape(1, -1)[0])))/len(list(y_test.reshape(1, -1)[0]))
```

```python
sum(abs(a-b) <=2 for a,b in zip(model.predict(X_test),list(y_test.reshape(1, -1)[0])))/len(list(y_test.reshape(1, -1)[0]))
```

```python
round_predict = [ round(elem) for elem in model.predict(X_test) ]
round_predict
```

```python
sum(abs(a-b) <=1 for a,b in zip(round_predict,list(y_test.reshape(1, -1)[0])))/len(list(y_test.reshape(1, -1)[0]))
```

We have a very good score but we can see that the predicted values are only 1 and 2 (when rounded)
