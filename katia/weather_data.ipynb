```python
import pandas as pd
weather = pd.read_csv("weather_2018+2019_upscale.csv",header=0,index_col=0)
```

```python
weather.drop(columns=["Year","m","d","Time"],inplace=True)
weather
```

```python
weather.Timestamp = pd.to_datetime(weather.Timestamp)
weather
```

```python
bikes_arrivals = pd.read_json("clean_arivals.json")
bikes_arrivals.rename(columns = {"datetime":"Timestamp"},inplace=True)
bikes_arrivals
```

Remove the November rows

```python
bikes_arrivals.drop(bikes_arrivals[pd.to_datetime(bikes_arrivals['Timestamp']).dt.month == 11].index, inplace=True)
bikes_arrivals
```

```python
#There is one malformed ID value, fix that
bikes_arrivals["id"].replace("19-", "19", inplace=True)
```

```python
#bikes_arrivals["Timestamp"] = bikes_arrivals["Timestamp"].astype(str)
#weather["Timestamp"] = weather["Timestamp"].astype(str)
final_arrivals_data = bikes_arrivals.merge(weather,on = "Timestamp",how="left")
final_arrivals_data
```

```python
# Doing the same for departures

bikes_departures = pd.read_json("clean_departures.json")
bikes_departures.rename(columns = {"datetime":"Timestamp"},inplace=True)
final_departures_data = bikes_departures.merge(weather,on = "Timestamp",how="left")
final_departures_data
```

```python
final_arrivals_data.to_csv('final_arrivals_data.csv')
final_departures_data.to_csv('final_departures_data.csv')
```
